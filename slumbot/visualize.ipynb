Visualize the slumbot bot results.

```python
import joblib
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot(filename, title):
    plt.figure(figsize=(10, 5))
    raw_results = joblib.load(f"../results/{filename}_raw.joblib")
    baseline_results = joblib.load(f"../results/{filename}_baseline.joblib")
    raw_results = raw_results[:10000] # only plot first 10000 hands
    baseline_results = baseline_results[:10000]
    raw_results = np.array(raw_results, dtype=np.float64)
    baseline_results = np.array(baseline_results, dtype=np.float64)
    raw_results /= 10000.0 # convert in terms of BB/100
    baseline_results /= 10000.0 # convert in terms of BB/100
    plt.plot(raw_results)
    plt.plot(baseline_results)
    plt.legend(["Raw BB/100", "Baseline BB/100"])
    plt.title(title + f" (Raw {str(raw_results[-1])} BB/100)")
    plt.xlabel("Number of hands")
    plt.ylabel("Earnings (BB/100)")
    plt.show()
```

```python
plot('slumbot_strategy_0_all-in', "Strategy 0: All-in")
```
